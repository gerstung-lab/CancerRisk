```python
## Custom Modules
#=======================================================================================================================
import sys
import os 
import tqdm
import h5py

import time
import pickle
import numpy as np 
import pandas as pd 

import plotly.graph_objects as go 
import plotly.express as px

import matplotlib as mpl
import matplotlib.pyplot as plt

ROOT_DIR = '/content/drive/MyDrive/tmp/plots/'


dd_theta = pd.read_csv(ROOT_DIR + 'theta.csv')

disease_ref = np.load(ROOT_DIR + 'disease_codes.npy', allow_pickle=True)
disease_ref = disease_ref[:, 1]

events = ['oesophagus', 'stomach', 'colorectal', 'liver',
                'pancreas', 'lung', 'melanoma', 'breast', 'cervix_uteri',
                'corpus_uteri', 'ovary', 'prostate', 'testis', 'kidney',
                'bladder', 'brain', 'thyroid', 'non_hodgkin_lymphoma', 'multiple_myeloma', 'AML', 'other', 'death']


Events = ['Oesophagus', 'Stomach', 'Colorectal', 'Liver',
                'Pancreas', 'Lung', 'Melanoma', 'Breast', 'Cervix uteri',
                'Corpus Uteri', 'Ovary', 'Prostate', 'Testis', 'Kidney',
                'Bladder', 'Brain', 'Thyroid', 'Non Hodgkin Lymphoma', 'Multiple Myeloma', 'AML']

gene_names = np.asarray([jj+ii for jj in Events for ii in [' First Degree', ' All', ' Multiple', ' Early']])

bth_names = np.asarray(['Alcoholic', 'Smoker', 'High Blood Pressure', 'Low Blood Pressure', 'Height', 'Weight', 'Age at first Birth'])

names = np.concatenate((disease_ref, gene_names, bth_names))

colormap= np.asarray(['#1E90FF', '#BFEFFF', '#191970', '#87CEFA', '#008B8B', '#946448', '#421a01', '#6e0b3c', 
                '#9370DB', '#7A378B', '#CD6090', '#006400', '#5ebd70', '#f8d64f', '#EEAD0E', '#f8d6cf',
                '#CDCB50', '#CD6600', '#FF8C69', '#8f0000', '#b3b3b3', '#454545'])

for cc in range(22):
    print(cc, events[cc])

```

```python
theta = np.exp(np.asarray(dd_theta.iloc[:, np.arange(2, 62, 3)]))
theta_lower = np.exp(np.asarray(dd_theta.iloc[:, np.arange(3, 63, 3)]))
theta_upper = np.exp(np.asarray(dd_theta.iloc[:, np.arange(4, 64, 3)]))

y = np.arange(theta.shape[0])
```

```python
import plotly.graph_objects as go 
fig = go.Figure()
n=300
for cc in range(20):
  fig.add_trace(go.Scatter(
      x=theta[:, cc],
      y=names, 
      mode="markers",
      marker_color=colormap[cc], 
      name=Events[cc],
      customdata=pd.DataFrame({'l': theta_lower[:, cc], 
              'u': theta_upper[:, cc], 
              'e': np.repeat(Events[cc], theta.shape[0]) }),
      hovertemplate='<b>Cancer: </b>: %{customdata[2]}<br>' + 
                    '<b>Var: </b>: %{y}<br>' + 
                    '<b>HR</b>: %{x:.2f} [%{customdata[0]:.2f}, %{customdata[1]:.2f}]<br>' + 
                    '<extra></extra>'
  ))

fig.update_xaxes(title_text="Hazard Ratio ", type="log", gridcolor='#acacad')
fig.update_yaxes(gridcolor='#acacad')


fig.update_layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
    width=1200,
    height=950,
    
)

fig.write_html(ROOT_DIR + 'dexp.html')
```
